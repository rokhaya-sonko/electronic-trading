#### Computing the Mid-Price

To build execution algorithms, we use the **mid-price**, which represents the center
of the high‚Äìlow range for each bar.

We compute it using the standard formula:

$$
\text{mid\_price} = \frac{\text{High} + \text{Low}}{2}
$$

This mid-price is a neutral estimate of the fair price inside the bar,
and it will be used throughout the execution study:

- to compute Market VWAP
- to simulate execution prices (mid ¬± spread)
- to evaluate execution costs (slippage, IS)

The final `market_data` table will contain only the fields required for execution:

- `minute` ‚Äî discrete time bucket
- `mid_price` ‚Äî computed mid-price
- `volume` ‚Äî traded volume



```python
# Choose the ticker and period
ticker = "AAPL" 
interval = "1m"
period = "1d" 

# Download OHLCV intraday data
raw = yf.download(
 ticker,
 interval=interval,
 period=period
)

# Reset index so that the timestamp becomes a normal column
df = raw.reset_index()

# Create a discrete time index (minute buckets)
df["minute"] = range(len(df))

# Compute the mid-price from High and Low
df["mid_price"] = (df["High"] + df["Low"]) / 2

# Use the traded volume as-is
df["volume"] = df["Volume"]

# Final market data for execution algorithms
market_data = df[["minute", "mid_price", "volume"]]

market_data.head()
```

## üìù Defining the client order

We now define a simple client order that we want to execute using different
execution strategies.

For this study, we assume:

- **Side**: BUY
- **Total quantity**: 100,000 shares
- **Time horizon**: from the first to the last minute in `market_data`

All execution algorithms (TWAP, VWAP, POV, IS-style) will try to execute
this same order over the same time window. This allows us to compare their
performance on a like-for-like basis.




```python
# üìù Define the client order

class ClientOrder:
    def __init__(self, side: str, quantity: float, start_minute: int, end_minute: int):
        assert side in ("BUY", "SELL")
        self.side = side
        self.quantity = quantity
        self.start_minute = start_minute
        self.end_minute = end_minute

    def __repr__(self):
        return (
                f"ClientOrder(side={self.side}, quantity={self.quantity}, "
                f"start={self.start_minute}, end={self.end_minute})"
                )
```

```python
start_minute = int(market_data["minute"].min())
end_minute = int(market_data["minute"].max())

client_order = ClientOrder(
               side="BUY",
               quantity=100_000,
               start_minute=start_minute,
               end_minute=end_minute)
client_order
```

## ‚öôÔ∏è Execution algorithms

We now define the execution strategies that will be used to schedule the client order. We work on a discrete time grid \( t = 0,.., T-1 \), where each bucket corresponds to one minute in `market_data`.
Let :
- \( Q \) = total order quantity (signed: > 0 for BUY, < 0 for SELL)
- \( V_t \) = market traded volume in bucket \( t \)
- \( w_t \) = weight applied to each bucket



- **TWAP ‚Äì Time-Weighted Average Price**

TWAP executes a constant quantity in each time bucket : $q_t = \frac{Q}{T}$. It ignores the intraday volume profile and focuses only on time.

- **VWAP ‚Äì Volume-Weighted Average Price**

VWAP tries to follow the market volume profile:  $ q_t = Q \cdot \frac{V_t}{\sum_{s=0}^{T-1} V_s} $. This allocates more quantity to buckets where the market trades more.


- **POV ‚Äì Participation of Volume**

POV executes a fixed participation rate ( $\alpha$ ) of the market volume: $ q_t^{\text{raw}} = \alpha \cdot V_t $ .  We then rescale ( $q_t^{\text{raw}}$) so that the total quantity matches ( $Q$ ).  This makes the execution adaptive to real-time market activity.


- **IS-style ‚Äì Implementation Shortfall style schedule**

An IS-style schedule is often more **front-loaded for a BUY order**
(to reduce the risk of the price drifting up), or **back-loaded for a SELL**.

We model a simple exponentially decaying profile for a BUY : $ w_t \propto e^{-k t} $ where ( $k > 0$ ) controls how aggressive the front-loading is and then normalize: $ q_t = Q \cdot \frac{w_t}{\sum_s w_s} $ .    This is a simplistic proxy for risk-aware execution.



```python
# Build execution schedules: TWAP, VWAP, POV, IS-style

def build_twap_schedule(order, market_data):
    """Equal quantity in each time bucket."""
    mask = (market_data["minute"] >= order.start_minute) & (market_data["minute"] <= order.end_minute)
    df = market_data.loc[mask].copy()
    T = len(df)
    sign = 1 if order.side == "BUY" else -1
    per_bucket = order.quantity / T
    df["target_qty"] = sign * per_bucket
    return df[["minute", "mid_price", "volume", "target_qty"]]


def build_vwap_schedule(order, market_data):
    """Quantity proportional to market volume in each bucket."""
    mask = (market_data["minute"] >= order.start_minute) & (market_data["minute"] <= order.end_minute)
    df = market_data.loc[mask].copy()
    total_vol = df["volume"].sum()
    sign = 1 if order.side == "BUY" else -1
    df["target_qty"] = sign * order.quantity * (df["volume"] / total_vol)
    return df[["minute", "mid_price", "volume", "target_qty"]]


def build_pov_schedule(order, market_data, participation_rate=0.1):
    """
    - trade participation_rate * volume in each bucket
    - then rescale so that the total absolute quantity = order.quantity
    """
    mask = (market_data["minute"] >= order.start_minute) & (market_data["minute"] <= order.end_minute)
    df = market_data.loc[mask].copy()
    raw_qty = participation_rate * df["volume"].values
    raw_total = raw_qty.sum()
    
    if raw_total == 0:
        scaled_qty = np.zeros_like(raw_qty)
    else:
        scaled_qty = raw_qty * (order.quantity / raw_total)
    
    sign = 1 if order.side == "BUY" else -1
    df["target_qty"] = sign * scaled_qty
    
    return df[["minute", "mid_price", "volume", "target_qty"]]


def build_is_style_schedule(order, market_data, k=0.05):
    """
    Simple Implementation Shortfall‚Äìstyle schedule.
    
    For a BUY: front-load the schedule with exponential decay.
    For a SELL: back-load the schedule (reverse time).
     """
    mask = (market_data["minute"] >= order.start_minute) & (market_data["minute"] <= order.end_minute)
    df = market_data.loc[mask].copy()

    minutes = df["minute"].values
    sign = 1 if order.side == "BUY" else -1

    if order.side == "BUY":
        t = minutes - minutes.min() # 0,1,2,...
        weights = np.exp(-k * t)
    else:
        t = minutes.max() - minutes # reversed
        weights = np.exp(-k * t)

    weights = weights / weights.sum()
    df["target_qty"] = sign * order.quantity * weights
    
    return df[["minute", "mid_price", "volume", "target_qty"]]
```

```python
# Build all four schedules
twap_sched = build_twap_schedule(client_order, market_data)
vwap_sched = build_vwap_schedule(client_order, market_data)
pov_sched = build_pov_schedule(client_order, market_data, participation_rate=0.12)
is_sched = build_is_style_schedule(client_order, market_data, k=0.08)

print('TWAP Schedule -\n ')
display(twap_sched.head())

print('VWAP Schedule -\n ')
display(vwap_sched.head())
```

### Execution Simulator (mid ¬± spread)

Once the execution schedules are generated, we need to convert them into actual *executed trades*.

A marketable BUY order executes near the **ask**,and a marketable SELL near the **bid**.
The mid-price represents the midpoint: $mid = \frac{bid + ask}{2}$
The bid‚Äìask spread is the transaction cost required to cross the market.

To approximate this microstructure in a simple and realistic way, we simulate execution at:

- **BUY** ‚Üí mid + half-spread
- **SELL** ‚Üí mid ‚Äì half-spread

Expressed in basis points: $p_t^{exec} = mid_t \times \bigl(1 + sign \times \frac{spread\_bps}{10\,000}\bigr)$
where:

- `sign = +1` for BUY
- `sign = -1` for SELL
- `spread_bps` is the assumed spread in basis points
(e.g. 1 bp = 0.01%)

The simulator produces, for each minute:

- the executed price
- the executed quantity
- the notional (price √ó quantity)

This output will then be used to compute all execution-quality metrics:
Execution VWAP, slippage, Implementation Shortfall, participation,
and completion profiles.

```python
#  Simple execution simulator (mid ¬± spread)

def simulate_execution(schedule, spread_bps=1.0):
    """
    Convert a schedule (target quantity per minute) into actual executed trades.
    
    Assumptions:
    - full fill at each bucket
    - execution at mid ¬± spread (mid + spread for BUY, mid - spread for SELL)
    - spread given in basis points
    """

    df = schedule.copy()

    # sign = +1 for BUY, -1 for SELL
    sign = np.sign(df["target_qty"])
    
    # Execution price = mid ¬± (spread in bps)
    df["exec_price"] = df["mid_price"] * (1 + sign * spread_bps / 10_000)
    
    # Execution notional = exec_price √ó target_qty
    df["exec_notional"] = df["exec_price"] * df["target_qty"]
    
    return df[
    ["minute", "mid_price", "volume", "target_qty", "exec_price", "exec_notional"]
    ]

# Run simulator for all algorithms
twap_exec = simulate_execution(twap_sched, spread_bps=1.0)
vwap_exec = simulate_execution(vwap_sched, spread_bps=1.0)
pov_exec = simulate_execution(pov_sched, spread_bps=1.0)
is_exec = simulate_execution(is_sched, spread_bps=1.0)

twap_exec.head()
```

#### üìè Execution Metrics

Once the simulated trades are generated for each algorithm, we compute some execution-quality metrics : 

- **Market VWAP** : the benchmark VWAP of the market over the trading horizon:  $ \text{VWAP}_{mkt} =
\frac{\sum_t mid_t \cdot volume_t}{\sum_t volume_t}
$

- **Execution VWAP** : the VWAP actually achieved by the algorithm: $ \text{VWAP}_{exec} =
\frac{\sum_t exec\_price_t \cdot qty_t}{\sum_t qty_t}
$

- **Slippage vs Market VWAP (bps)** : for a BUY - $ slippage_{VWAP} = 10^4 \cdot
\frac{VWAP_{exec} - VWAP_{mkt}}{VWAP_{mkt}}$

- **Slippage vs Arrival Price (bps)** :  arrival Price (AP) is the mid-price at the moment the order starts.

- **Implementation Shortfall (IS)** : true economic cost of execution: $IS =
\frac{\sum_t exec\_price_t \cdot qty_t - Q \cdot AP}
{Q \cdot AP}
$

- **Participation** : ratio of executed quantity to market volume: $part_t = \frac{|qty_t|}{volume_t}$

- **Completion profile** : tracks how quickly the order is completed over time.


```python
def market_vwap(market_data, start, end):
    df = market_data[(market_data["minute"] >= start) & (market_data["minute"] <= end)]
    return (df["mid_price"] * df["volume"]).sum() / df["volume"].sum()

def execution_vwap(exec_df):
    return exec_df["exec_notional"].sum() / exec_df["target_qty"].sum()

def slippage_vs_vwap(exec_vwap, mkt_vwap, side="BUY"):
    if side == "BUY":
        return 10_000 * (exec_vwap - mkt_vwap) / mkt_vwap
    else:
        return 10_000 * (mkt_vwap - exec_vwap) / mkt_vwap

def arrival_price(market_data, minute):
    return market_data.loc[market_data["minute"] == minute, "mid_price"].iloc[0]

def implementation_shortfall(exec_df, AP, side="BUY"):
    Q = exec_df["target_qty"].sum()
    notional = exec_df["exec_notional"].sum()

    if side == "BUY":
        return (notional - Q * AP) / (Q * AP)
    else:
        return (Q * AP - notional) / (Q * AP)

def average_participation(exec_df, market_data):
    df = exec_df.merge(market_data[["minute", "volume"]], on="minute")
    df["part"] = df["target_qty"].abs() / df["volume"]
    return df["part"].mean()

def completion_profile(exec_df):
    df = exec_df.copy()
    df["cum_qty"] = df["target_qty"].cumsum()
    df["completion"] = df["cum_qty"] / df["target_qty"].sum()
    return df
```

```python
start = client_order.start_minute
end = client_order.end_minute
AP = arrival_price(market_data, start)

mkt_vwap_val = market_vwap(market_data, start, end)

results = []

for name, df_exec in [ ("TWAP", twap_exec), ("VWAP", vwap_exec),("POV", pov_exec), ("IS", is_exec)]:
    exec_vwap_val = execution_vwap(df_exec)
    slip_vwap = slippage_vs_vwap(exec_vwap_val, mkt_vwap_val, side=client_order.side)
    slip_AP = slippage_vs_vwap(exec_vwap_val, AP, side=client_order.side)
    IS = implementation_shortfall(df_exec, AP, side=client_order.side)
    part = average_participation(df_exec, market_data)
 
results.append({
 "algo": name,
 "mkt_vwap": mkt_vwap_val,
 "exec_vwap": exec_vwap_val,
 "slippage_vs_vwap_bps": slip_vwap,
 "slippage_vs_AP_bps": slip_AP,
 "implementation_shortfall": IS,
 "avg_participation": part
 })

pd.DataFrame(results)
```
