We use the `yfinance` library to download real intraday market data.

#### Computing the Mid-Price

To build execution algorithms, we use the **mid-price**, which represents the center
of the high‚Äìlow range for each bar.

We compute it using the standard formula:

$$
\text{mid\_price} = \frac{\text{High} + \text{Low}}{2}
$$

This mid-price is a neutral estimate of the fair price inside the bar,
and it will be used throughout the execution study:

- to compute Market VWAP
- to simulate execution prices (mid ¬± spread)
- to evaluate execution costs (slippage, IS)

The final `market_data` table will contain only the fields required for execution:

- `minute` ‚Äî discrete time bucket
- `mid_price` ‚Äî computed mid-price
- `volume` ‚Äî traded volume



```python
# Choose the ticker and period
ticker = "AAPL" 
interval = "1m"
period = "1d" 

# Download OHLCV intraday data
raw = yf.download(
 ticker,
 interval=interval,
 period=period
)

# Reset index so that the timestamp becomes a normal column
df = raw.reset_index()

# Create a discrete time index (minute buckets)
df["minute"] = range(len(df))

# Compute the mid-price from High and Low
df["mid_price"] = (df["High"] + df["Low"]) / 2

# Use the traded volume as-is
df["volume"] = df["Volume"]

# Final market data for execution algorithms
market_data = df[["minute", "mid_price", "volume"]]

market_data.head()
```

## üìù Defining the client order

We now define a simple client order that we want to execute using different
execution strategies.

For this study, we assume:

- **Side**: BUY
- **Total quantity**: 100,000 shares
- **Time horizon**: from the first to the last minute in `market_data`

All execution algorithms (TWAP, VWAP, POV, IS-style) will try to execute
this same order over the same time window. This allows us to compare their
performance on a like-for-like basis.




```python
# üìù Define the client order

class ClientOrder:
    def __init__(self, side: str, quantity: float, start_minute: int, end_minute: int):
        assert side in ("BUY", "SELL")
        self.side = side
        self.quantity = quantity
        self.start_minute = start_minute
        self.end_minute = end_minute

    def __repr__(self):
        return (
                f"ClientOrder(side={self.side}, quantity={self.quantity}, "
                f"start={self.start_minute}, end={self.end_minute})"
                )
```

```python
start_minute = int(market_data["minute"].min())
end_minute = int(market_data["minute"].max())

client_order = ClientOrder(
               side="BUY",
               quantity=100_000,
               start_minute=start_minute,
               end_minute=end_minute)
client_order
```

## ‚öôÔ∏è Execution algorithms

We now define the execution strategies that will be used to schedule the client order. We work on a discrete time grid \( t = 0,.., T-1 \), where each bucket corresponds to one minute in `market_data`.
Let :
> - \( Q \) = total order quantity (signed: > 0 for BUY, < 0 for SELL)
> - \( V_t \) = market traded volume in bucket \( t \)
> - \( w_t \) = weight applied to each bucket



- **TWAP ‚Äì Time-Weighted Average Price**

> TWAP executes a constant quantity in each time bucket : $q_t = \frac{Q}{T}$. It ignores the intraday volume profile and focuses only on time.

- **VWAP ‚Äì Volume-Weighted Average Price**

> VWAP tries to follow the market volume profile:  $ q_t = Q \cdot \frac{V_t}{\sum_{s=0}^{T-1} V_s} $. This allocates more quantity to buckets where the market trades more.


- **POV ‚Äì Participation of Volume**

> POV executes a fixed participation rate ( $\alpha$ ) of the market volume: $ q_t^{\text{raw}} = \alpha \cdot V_t $ .  We then rescale ( $q_t^{\text{raw}}$) so that the total quantity matches ( $Q$ ).  This makes the execution adaptive to real-time market activity.


- **IS-style ‚Äì Implementation Shortfall style schedule**

> An IS-style schedule is often more **front-loaded for a BUY order**
(to reduce the risk of the price drifting up), or **back-loaded for a SELL**.
We model a simple exponentially decaying profile for a BUY : $ w_t \propto e^{-k t} $ where ( $k > 0$ ) controls how aggressive the front-loading is and then normalize: $ q_t = Q \cdot \frac{w_t}{\sum_s w_s} $ .
This is a simplistic proxy for risk-aware execution.



```python
# Build execution schedules: TWAP, VWAP, POV, IS-style

def build_twap_schedule(order, market_data):
    """Equal quantity in each time bucket."""
    mask = (market_data["minute"] >= order.start_minute) & (market_data["minute"] <= order.end_minute)
    df = market_data.loc[mask].copy()
    T = len(df)
    sign = 1 if order.side == "BUY" else -1
    per_bucket = order.quantity / T
    df["target_qty"] = sign * per_bucket
    return df[["minute", "mid_price", "volume", "target_qty"]]


def build_vwap_schedule(order, market_data):
    """Quantity proportional to market volume in each bucket."""
    mask = (market_data["minute"] >= order.start_minute) & (market_data["minute"] <= order.end_minute)
    df = market_data.loc[mask].copy()
    total_vol = df["volume"].sum()
    sign = 1 if order.side == "BUY" else -1
    df["target_qty"] = sign * order.quantity * (df["volume"] / total_vol)
    return df[["minute", "mid_price", "volume", "target_qty"]]


def build_pov_schedule(order, market_data, participation_rate=0.1):
    """
    - trade participation_rate * volume in each bucket
    - then rescale so that the total absolute quantity = order.quantity
    """
    mask = (market_data["minute"] >= order.start_minute) & (market_data["minute"] <= order.end_minute)
    df = market_data.loc[mask].copy()
    raw_qty = participation_rate * df["volume"].values
    raw_total = raw_qty.sum()
    
    if raw_total == 0:
        scaled_qty = np.zeros_like(raw_qty)
    else:
        scaled_qty = raw_qty * (order.quantity / raw_total)
    
    sign = 1 if order.side == "BUY" else -1
    df["target_qty"] = sign * scaled_qty
    
    return df[["minute", "mid_price", "volume", "target_qty"]]


def build_is_style_schedule(order, market_data, k=0.05):
    """
    Simple Implementation Shortfall‚Äìstyle schedule.
    
    For a BUY: front-load the schedule with exponential decay.
    For a SELL: back-load the schedule (reverse time).
     """
    mask = (market_data["minute"] >= order.start_minute) & (market_data["minute"] <= order.end_minute)
    df = market_data.loc[mask].copy()

    minutes = df["minute"].values
    sign = 1 if order.side == "BUY" else -1

    if order.side == "BUY":
        t = minutes - minutes.min() # 0,1,2,...
        weights = np.exp(-k * t)
    else:
        t = minutes.max() - minutes # reversed
        weights = np.exp(-k * t)

    weights = weights / weights.sum()
    df["target_qty"] = sign * order.quantity * weights
    
    return df[["minute", "mid_price", "volume", "target_qty"]]
```

```python
# Build all four schedules
twap_sched = build_twap_schedule(client_order, market_data)
vwap_sched = build_vwap_schedule(client_order, market_data)
pov_sched = build_pov_schedule(client_order, market_data, participation_rate=0.12)
is_sched = build_is_style_schedule(client_order, market_data, k=0.08)

print('TWAP Schedule -\n ')
display(twap_sched.head())

print('VWAP Schedule -\n ')
display(vwap_sched.head())
```

### Execution Simulator (mid ¬± spread)

Once the execution schedules are generated, we need to convert them into actual *executed trades*.

A marketable BUY order executes near the **ask**,and a marketable SELL near the **bid**.
The mid-price represents the midpoint: $mid = \frac{bid + ask}{2}$
The bid‚Äìask spread is the transaction cost required to cross the market.

To approximate this microstructure in a simple and realistic way, we simulate execution at:

> - **BUY** ‚Üí mid + half-spread
> - **SELL** ‚Üí mid ‚Äì half-spread

Expressed in basis points: $p_t^{exec} = mid_t \times \bigl(1 + sign \times \frac{spread\_bps}{10\,000}\bigr)$
where:

> - `sign = +1` for BUY
> - `sign = -1` for SELL
> - `spread_bps` is the assumed spread in basis points
(e.g. 1 bp = 0.01%)

The simulator produces, for each minute:

- the executed price
- the executed quantity
- the notional (price √ó quantity)

This output will then be used to compute all execution-quality metrics:
Execution VWAP, slippage, Implementation Shortfall, participation,
and completion profiles.

```python
#  Simple execution simulator (mid ¬± spread)

def simulate_execution(schedule, spread_bps=1.0):
    """
    Convert a schedule (target quantity per minute) into actual executed trades.
    
    Assumptions:
    - full fill at each bucket
    - execution at mid ¬± spread (mid + spread for BUY, mid - spread for SELL)
    - spread given in basis points
    """

    df = schedule.copy()

    sign = np.sign(df["target_qty"])
    df["exec_price"] = df["mid_price"] * (1 + sign * spread_bps / 10_000)
    df["exec_notional"] = df["exec_price"] * df["target_qty"]

    df = df.set_index("minute")

    return df[["mid_price", "volume", "target_qty", "exec_price", "exec_notional"]]

# Run simulator for all algorithms
twap_exec = simulate_execution(twap_sched, spread_bps=1.0)
vwap_exec = simulate_execution(vwap_sched, spread_bps=1.0)
pov_exec = simulate_execution(pov_sched, spread_bps=1.0)
is_exec = simulate_execution(is_sched, spread_bps=1.0)

twap_exec.head()
```

### üìä Execution Metrics

This notebook evaluates execution quality using standard metrics employed on modern electronic execution desks. Each metric captures a different execution dimension: cost, benchmark performance, timing quality, market impact, or liquidity interaction.

We first define all metrics (with formulas and interpretation). Then we map **which metrics are used for each algorithm** ‚Äî and *why*.

- **Execution VWAP** : Average execution price across all fills. Baseline for comparing strategies.
  $ VWAP_{exec} = \frac{\sum p_t q_t}{\sum q_t} $

```python
def execution_vwap(exec_df):
 df = flatten(exec_df)
 return (df["exec_price"] * df["target_qty"]).sum() / df["target_qty"].sum()
```

- **Market VWAP** : Benchmark computed from market trades. Reference for VWAP-based trading.
$ VWAP_{mkt} = \frac{\sum p_t^{mkt} v_t^{mkt}}{\sum v_t^{mkt}}$

```python
def market_vwap(mkt_df, start, end):
 df = mkt_df[(mkt_df["minute"] >= start) & (mkt_df["minute"] <= end)]
 return (df["mid_price"] * df["volume"]).sum() / df["volume"].sum()
```

- **iVWAP Slippage (bps)** :
$slip_{VWAP} = 10^4 \cdot \frac{VWAP_{exec} - VWAP_{mkt}}{VWAP_{mkt}}$


> - Negative ‚Üí beat the market 
> - Positive ‚Üí underperformed 
> - Primary metric for VWAP strategies

```python
def ivwap_slippage(exec_df, mkt_df, start, end, side="BUY"):
    """
    Compute Implementation VWAP Slippage (in basis points).

    Parameters
    ----------
    exec_df : pd.DataFrame
        Executed trades dataframe.
    mkt_df : pd.DataFrame
        Market data (prices, volumes‚Ä¶).
    start : datetime
        Start time for market VWAP.
    end : datetime
        End time for market VWAP.
    side : str, optional
        "BUY" or "SELL". Default is "BUY".
    Returns
    -------
    float
        IVWAP slippage in basis points (bps).
    """
    exec_vwap = execution_vwap(exec_df)
    mkt_vwap = market_vwap(mkt_df, start, end)

    if side.upper() == "BUY":
        return 1e4 * (exec_vwap - mkt_vwap) / mkt_vwap

    # SELL: slippage flips sign
    return 1e4 * (mkt_vwap - exec_vwap) / mkt_vwap
```

- **Arrival Price (AP)** : Neutral decision price when the order arrived.
$ AP = mid_{start} $. 

```python
def arrival_price(mkt_df, start_minute):
 return mkt_df.loc[mkt_df["minute"] == start_minute, "mid_price"].iloc[0]
```

- **AP Slippage (bps)** : Measures cost relative to the decision price. Used when the algorithm has no benchmark (TWAP, POV, IS-style). $slip_{AP} = 10^4 \cdot \frac{VWAP_{exec} - AP}{AP}$. 


```python
def ap_slippage(exec_df, mkt_df, start, side="BUY"):
    exec_vwap = execution_vwap(exec_df)
    AP = arrival_price(mkt_df, start)
    if side == "BUY":
        return 1e4 * (exec_vwap - AP) / AP
    else:
        return 1e4 * (AP - exec_vwap) / AP
```

- **Implementation Shortfall**: Most complete cost metric (drift + spread + timing + market impact) : $IS = \frac{\sum p_t q_t - Q \cdot AP}{Q \cdot AP}$

```python
def implementation_shortfall(exec_df, mkt_df, start, side="BUY"):
    df = flatten(exec_df)
    AP = arrival_price(mkt_df, start)
    notional = (df["exec_price"] * df["target_qty"]).sum()
    Q = df["target_qty"].sum()
    if side == "BUY":
        return (notional - Q * AP) / (Q * AP)
    else:
        return (Q * AP - notional) / (Q * AP)
```

- **Participation Rate** : Aggressiveness and market impact exposure. Primary for POV. $part_t = \frac{|q_t|}{volume_t^{mkt}}$. 

```python
def participation_rate(exec_df, mkt_df):
    df = flatten(exec_df)
    mkt = mkt_df[["minute", "volume"]].rename(columns={"volume": "mkt_volume"})
    merged = df.merge(mkt, on="minute", how="left")
    merged["participation"] = merged["target_qty"].abs() / merged["mkt_volume"]
    return merged["participation"].mean()
```

- **Adverse Selection** : Critical for POV and IS-style strategies.
$AS_t =\begin{cases}mid_{t+h} - p_t & \text{(buy)} \\p_t - mid_{t+h} & \text{(sell)}\end{cases}$.
> - Positive ‚Üí good timing 
> - Negative ‚Üí poor timing 

```python
def adverse_selection(exec_df, mkt_df, horizon=1):
    df = flatten(exec_df)
    mkt = mkt_df[["minute", "mid_price"]].copy()
    mkt["mid_after"] = mkt["mid_price"].shift(-horizon)
    
    merged = df.merge(mkt, on="minute", how="left")
    sign = np.sign(merged["target_qty"]) # +1 buy | -1 sell
    
    merged["AS"] = np.where( sign > 0, merged["mid_after"] - merged["exec_price"], # BUY 
                             merged["exec_price"] - merged["mid_after"] # SELL
                            )
    return merged["AS"].mean()
```

 - **Completion Profile** : Shape reveals the strategy‚Äôs pacing:
  $ C_t = \frac{\sum_{k \le t} q_k}{Q} $ . 
> - TWAP ‚Üí linear 
> - VWAP ‚Üí follows liquidity curve 
> - IS-style ‚Üí front-loaded or back-loaded 

```python
def completion_profile(exec_df):
 df = flatten(exec_df).sort_values("minute").copy()
 df["cum_qty"] = df["target_qty"].cumsum()
 df["completion"] = df["cum_qty"] / df["target_qty"].sum()
 return df
```

| Algorithm | Primary Metric | Secondary Metrics |
|----------|----------------|-------------------|
| **TWAP** | AP Slippage | IS, Completion, Participation, AS |
| **VWAP** | iVWAP Slippage | IS, Participation, AS, Completion |

#### Execution Analysis

```python
class ExecutionAnalysis:
    """
    Wrapper class to compute execution metrics for multiple algorithms.
    """

    def __init__(self, market_data, client_order):
        self.client_order = client_order
        if isinstance(market_data.columns, pd.MultiIndex):
            market_data = market_data.copy()
            market_data.columns = [
                "_".join([str(c) for c in col]).strip("_")
                for col in market_data.columns
            ]

        self.market_data = market_data


    # -----------------------------------------------------------
    # Compute metrics for a single algorithm
    # -----------------------------------------------------------
    def compute_metrics(self, name, exec_df):
        start = self.client_order.start_minute
        end = self.client_order.end_minute
        side = self.client_order.side
        
        exec_df = exec_df.reset_index()


        comp = completion_profile(exec_df)
        mid_idx = len(comp) // 2

        return {
            "algo": name,
            "execution_vwap": execution_vwap(exec_df),
            "ivwap_slippage_bps": ivwap_slippage(exec_df, self.market_data, start, end, side),
            "ap_slippage_bps": ap_slippage(exec_df, self.market_data, start, side),
            "implementation_shortfall": implementation_shortfall(exec_df, self.market_data, start, side),
            "avg_participation": participation_rate(exec_df, self.market_data),
            "avg_adverse_selection": adverse_selection(exec_df, self.market_data),
            "completion_mid": comp["completion"].iloc[mid_idx],
            "completion_final": comp["completion"].iloc[-1],
        }

    # -----------------------------------------------------------
    # Compute metrics for all algorithms
    # -----------------------------------------------------------
    def compute_all(self, twap, vwap, pov, is_exec):
        rows = [
            self.compute_metrics("TWAP", twap),
            self.compute_metrics("VWAP", vwap),
            self.compute_metrics("POV", pov),
            self.compute_metrics("IS-style", is_exec),
        ]
        return pd.DataFrame(rows)

analysis = ExecutionAnalysis(market_data, client_order)
algo_metrics_df = analysis.compute_all(twap_exec,vwap_exec,pov_exec,is_exec)
algo_metrics_df
```

#### Plot metrics

```python
def plot_interactive_metrics(df):
    # ---- Bar charts for slippages / IS / participation -------------------
    fig = make_subplots(rows=2, cols=2, subplot_titles=[
        "iVWAP Slippage (bps)",
        "AP Slippage (bps)",
        "Implementation Shortfall",
        "Avg Participation"
    ])

    metrics = [
        "ivwap_slippage_bps",
        "ap_slippage_bps",
        "implementation_shortfall",
        "avg_participation"
    ]

    row_col = [(1,1), (1,2), (2,1), (2,2)]

    for (row, col), metric in zip(row_col, metrics):
        fig.add_trace(
            go.Bar(
                x=df["algo"],
                y=df[metric],
                text=df[metric].round(4),
                textposition="outside"
            ),
            row=row, col=col
        )

    fig.update_layout(height=750, width=900, showlegend=False)
    fig.show()
```

```python
plot_interactive_metrics(algo_metrics_df)
```
